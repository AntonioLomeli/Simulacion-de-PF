Obtén los datos de las siguientes empresas desde el 1 de enero de 2020 hasta hoy:


Tesla

```python
import pandas as pd
import yfinance as yf 
import matplotlib.pyplot as plt
```

```python
data = yf.download(['TSLA','BAC','INTC','MSFT','AAPL'], start='2020-01-01')
data.head()
```

Grafica sus precios de cierre ajustado:  Adj Close  

```python
plt.figure(figsize=(12,6))
plt.plot(data['Adj Close'], label = ['TSLA','BAC','INTC','MSFT','AAPL'])
plt.title('Precio de cierre Ajustado de Acciones')
plt.legend()
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre ($)')
plt.grid(True)
plt.show()
```

Obtén el precio de cierre ajustado promedio para cada empresa.

```python
data['Adj Close'].mean()
```

Encuentra los días en los que Apple ha tenido un precio de cierre ajustado mayor a 180.

```python
dias_apple = pd.DataFrame(data[data['Adj Close']['AAPL'] > 180]['Adj Close']['AAPL'])
dias_apple
```
